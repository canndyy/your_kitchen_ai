```python
import requests
import numpy as np
import pandas as pd
from io import StringIO, BytesIO
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
url = 'http://127.0.0.1:8003'
```

```python
filename = 'kitchen_test.jpeg'
```

```python
img = Image.open(filename)
img
```

```python
arr=np.array(img)
img_shape=arr.shape
dtype_arr=arr.dtype
byte_arr=arr.tobytes()
```

```python
from_bytes = np.frombuffer(byte_arr, dtype = arr.dtype)
files = {'my_file': byte_arr}
```

```python
response = requests.post(url + '/file',files=files,data={'shape': str(img_shape), 'dtype': str(dtype_arr)})
```
