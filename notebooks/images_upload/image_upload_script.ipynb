**Required ingredients:**

```python
kaggle_vi_ingredients = ['Broccoli','Cabbage','Carrot',\
                         'Cauliflower','Cucumber','Potato','Pumpkin','Tomato']
```

```python
client = storage.Client.from_service_account_json(json_credentials_path=os.environ['GOOGLE_APPLICATION_CREDENTIALS'])
bucket = client.get_bucket('kitchen-ai-images')
```

```python
def upload_kaggle_vi_imgs():
    filename_tuples = []
    for ingredient in kaggle_vi_ingredients:
        index = 0
        for source in ['test','train','validation']:

            for filename in os.listdir(f'../../data/Vegetable Images/{source}/{ingredient}/'):

                f = os.path.join(f'../../data/Vegetable Images/{source}/{ingredient}/', f"{filename}")
                if os.path.isfile(f):
                    gcs_object = bucket.blob(f'{ingredient.upper()}_{index}.jpg')
                    gcs_object.upload_from_filename(os.path.join(f'../../data/Vegetable Images/{source}/{ingredient}/', f"{filename}"))

                    filename_tuples.append((filename, f'{ingredient.upper()}_vi_{index}.jpg'))
                    index += 1
                    print(f)
```

```python
upload_kaggle_vi_imgs()
```

To remove all files from bucket, run `gsutil -m rm 'gs://kitchen-ai-images/**`
